GRADIENT FOR LOSS FUNCTION

```python
%config InlineBackend.figure_format = 'retina'
import pandas as pd
import seaborn as sns
```

```python
#Task: minimize function y = (θ – 5)**2
#Find the value of θ, on which function gets minimum. Evident, that it is 0. Let's prove it.

n_iter = 20 # количество итераций
learning_rate = 0.2 # скорость обучения
def func(x):
    return (x - 5) ** 2
def func_derivative(x):
    return 2 * (x - 5)
previous_x, current_x = 0, 0

data1 = []
data2 = []
datai = []
df = pd.DataFrame(columns = ['function','derivative', 'i'])

for i in range(n_iter):
    current_x = previous_x - learning_rate * func_derivative(previous_x)
    previous_x = current_x
    data1.append(func(current_x))
    data2.append(func_derivative(current_x))
    datai.append(i)
    print("theta", format(current_x, ".6f"),
          "function value=", format(func(current_x), ".6f"),
          "derivative=", format(func_derivative(current_x), ".6f"))
    
import matplotlib.pyplot as plt
import seaborn as sns
df['function'] = data1
df['derivative'] = data2
df['i'] = datai
sns.set_style("darkgrid")
plt.figure(figsize=(12,5))
plt.title("Data")
sns.lineplot(data=df[['function','derivative']])
    
```
