```python
#query_res = pd.read_sql("SELECT City FROM AIRPORTS;", connection)
query_res3 = pd.read_sql("SELECT City, Segment, State FROM AIRPORTS WHERE Country='United States' AND Segment='Consumer';",connection)
query_res3
```

```python
query_res4 = pd.read_sql("SELECT City FROM AIRPORTS WHERE Segment='Consumer' ORDER BY 'Ship Date';", connection)
query_res4
```

```python
pandas_res4 = df[df['Segment'] == 'Consumer'].sort_values('Ship Date')
pandas_res4
```

```python
query_res5 = pd.read_sql("SELECT * FROM AIRPORTS WHERE Segment='Consumer' ORDER BY 'Ship Date' DESC;", connection)
query_res5
```
