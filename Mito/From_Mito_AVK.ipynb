```python
import pandas as pd

#1. Imported flight_delays_train.csv
flight_delays_train_3 = pd.read_csv('./flight_delays_train.csv')
```

```python
#2. Copied from Mito Jupyter-Lab:

#NB! To show the graph take away renderer="iframe"

import plotly.express as px

# Filter the dataframe so that it does not crash the browser
flight_delays_train_3_filtered = flight_delays_train_3.head(1000)

# Construct the graph and style it. Further customize your graph by editing this code.
# See Plotly Documentation for help: https://plotly.com/python/plotly-express/
fig = px.scatter(flight_delays_train_3_filtered, x=['DepTime', 'Distance'], y='dep_delayed_15min')
fig.update_layout(
    title='DepTime, Distance, dep_delayed_15min (first 1000 rows) scatter plot', 
    xaxis = dict(
        showgrid=True, 
        rangeslider = dict(
            visible=True, 
            thickness=0.05
        )
    ), 
    yaxis = dict(
        showgrid=True
    ), 
    legend = dict(
        orientation='v'
    ), 
    paper_bgcolor='#FFFFFF'
)

fig.show()
```

```python
# 3. Pivoting/Group By.
tmp_df = flight_delays_train_3[flight_delays_train_3['dep_delayed_15min'] == 'Y']
tmp_df = tmp_df[['Origin']].copy()
pivot_table = tmp_df.pivot_table(
    index=['Origin'],
    values=['Origin'],
    aggfunc={'Origin': ['count']}
)
#pivot_table = pivot_table.set_axis([flatten_column_header(col) for col in pivot_table.keys()], axis=1) ### gave an error
flight_delays_train_3_pivot = pivot_table.reset_index()
```
