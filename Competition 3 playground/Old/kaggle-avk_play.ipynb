```python
X = train.drop(['id'], axis = 1)
```

```python
X = X.drop(['cost'], axis = 1)
```

```python
X
```

```python
y = train['cost']
y
```

```python
#scaling X.
X = (X-X.min())/(X.max()-X.min())
X
```

```python
import seaborn as sns
corr_matrix = X.corr()
sns.heatmap(corr_matrix);
#Yes correlation, it depends
```

```python
# Drop correlated

X = X.drop(['salad_bar'], axis = 1)
X
```

```python
#Normalization of one feature

import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.histplot(X.gross_weight, kde=True)
ax.set_title("Input data before")
plt.show()
```

```python
#Normalization of gross_weight

from scipy import stats
from scipy.stats import boxcox
index_gross_weight = X.gross_weight > 0
positive_gross_weight = X.gross_weight.loc[index_gross_weight]
normalized_gross_weight = pd.Series(stats.boxcox(positive_gross_weight)[0], 
                               name='result', index=positive_gross_weight.index)

#X.gross_weight = normalized_pledges
ax = sns.histplot(normalized_gross_weight, kde=True)
ax.set_title("Input data after normalization")
plt.show()
#Result: use, difference exists!

X.gross_weight = normalized_gross_weight
```

```python
#Normalize all feartures with no 0/1:
from scipy import stats
from scipy.stats import boxcox

index_store_salesinmillions = X['store_sales(in millions)'] > 0
index_unit_salesinmillions = X['unit_sales(in millions)'] > 0
index_total_children  = X.total_children > 0
index_num_children_at_home = X.num_children_at_home > 0
index_avg_cars_athomeapprox = X['avg_cars_at home(approx).1'] > 0
index_gross_weight  = X.gross_weight > 0
index_units_per_case= X.units_per_case > 0
index_store_sqft = X.store_sqft > 0

positive_store_salesinmillions = X['store_sales(in millions)'].loc[index_store_salesinmillions]
positive_unit_salesinmillions = X['unit_sales(in millions)'].loc[index_unit_salesinmillions]
positive_total_children = X.total_children.loc[index_total_children]
positive_num_children_at_home = X.num_children_at_home.loc[index_num_children_at_home]
positive_avg_cars_athomeapprox = X['avg_cars_at home(approx).1'].loc[index_avg_cars_athomeapprox]
positive_gross_weight = X.gross_weight.loc[index_gross_weight]
positive_units_per_case = X.units_per_case.loc[index_units_per_case]
positive_store_sqft = X.store_sqft.loc[index_store_sqft]


normalized_store_salesinmillions = pd.Series(stats.boxcox(positive_store_salesinmillions)[0], 
                               name='result', index=positive_store_salesinmillions.index)
normalized_unit_salesinmillions = pd.Series(stats.boxcox(positive_unit_salesinmillions)[0], 
                               name='result', index=positive_unit_salesinmillions.index)
normalized_total_children = pd.Series(stats.boxcox(positive_total_children)[0], 
                               name='result', index=positive_total_children.index)
normalized_num_children_at_home = pd.Series(stats.boxcox(positive_num_children_at_home )[0], 
                               name='result', index=positive_num_children_at_home.index)
normalized_avg_cars_athomeapprox = pd.Series(stats.boxcox(positive_avg_cars_athomeapprox)[0], 
                               name='result', index=positive_avg_cars_athomeapprox.index)
normalized_gross_weight = pd.Series(stats.boxcox(positive_gross_weight)[0], 
                               name='result', index=positive_gross_weight.index)
normalized_units_per_case = pd.Series(stats.boxcox(positive_units_per_case)[0], 
                               name='result', index=positive_units_per_case.index)
normalized_store_sqft = pd.Series(stats.boxcox(positive_store_sqft)[0], 
                               name='result', index=positive_store_sqft.index)


X['store_sales(in millions)'] = normalized_store_salesinmillions 
X['unit_sales(in millions)'] = normalized_unit_salesinmillions   
X.total_children = normalized_total_children   
X.num_children_at_home = normalized_num_children_at_home 
X['avg_cars_at home(approx).1'] = normalized_avg_cars_athomeapprox 
X.gross_weight = normalized_gross_weight   
X.units_per_case  = normalized_units_per_case    
X.store_sqft  = normalized_store_sqft    


#X.gross_weight = normalized_pledges
ax = sns.histplot(normalized_gross_weight, kde=True)
ax.set_title("Input data after normalization")
plt.show()
#Result: use, difference exists!  
```
