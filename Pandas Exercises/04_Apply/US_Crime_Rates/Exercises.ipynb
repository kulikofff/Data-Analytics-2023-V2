##### Have you noticed that the type of Year is int64. But pandas has a different type to work with Time Series. Let's see it now.

### Step 5. Convert the type of the column Year to datetime64

```python
crime.Year = pd.to_datetime(crime['Year'], format = '%Y')
#crime.Year = pd.to_datetime(crime['Year'])
crime.head()
```

### Step 6. Set the Year column as the index of the dataframe

```python
crime = crime.set_index('Year')
crime.head()
```

### Step 7. Delete the Total column

```python
crime = crime.drop('Total', axis =1)
crime.head()

#or
#del crime['Total']
```

### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

```python
# To learn more about .resample (https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.resample.html)
# To learn more about Offset Aliases (http://pandas.pydata.org/pandas-docs/stable/timeseries.html#offset-aliases)

#crimedec = crime.groupby((crime.index.year//10)*10).sum() - works:)

crimedec = crime.resample('10AS').sum()

c = pd.DataFrame()
c = crime['Population'].resample('10AS').max()
crimedec['Population'] = c

crimedec
```

```python
#Version from author
crimedec.idxmax(0)
#Return index of first occurrence of maximum over requested axis. NA/null values are excluded.
```

### Step 9. What is the most dangerous decade to live in the US?

```python
#my version. On sum of crimes on a decade
crimesmax = crimedec.drop('Population', axis = 1)
#for i in crimesmax.index:
 #   print(crimesmax)
crimesmax['sum'] = crimesmax.sum(axis = 1)
    
crimesmax.sort_values(['sum'], ascending=False).iloc[[0]]
```
